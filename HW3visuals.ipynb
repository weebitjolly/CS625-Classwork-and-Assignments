# First Visual Idiom
## Stacked Barchart of Norfolk Water Quality

```python
waterquality = pd.read_csv('/content/drive/MyDrive/Water_Quality_Test_Results_20250223.csv')
waterquality.head(10)
```

```python
# title Regulation

from matplotlib import pyplot as plt
import seaborn as sns
waterquality.groupby('Regulation').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# creating a stacked barchart looking at the types of regulations tested at each location
stacked_data = waterquality.groupby(['Regulation', 'Treatment Plant']).size().unstack()
stacked_data.plot(kind='bar', stacked=True, color=sns.color_palette('pastel'))
plt.gca().spines[['top', 'right']].set_visible(False)
plt.xlabel("Regulation")
plt.ylabel("Count")
plt.title("Stacked Bar Chart of Water Quality by Regulation and Treatment Plant")
plt.legend(title='Treatment Plant')
plt.show()
```

```python
#line chart for amazon stocks
AMZN = pd.read_csv('/content/drive/MyDrive/AMZN.csv')
AMZN.head(10)
```

```python


import pandas as pd
import matplotlib.pyplot as plt
# Convert 'Date' column to datetime objects
AMZN['Date'] = pd.to_datetime(AMZN['Date'])

# Set 'Date' as the index
AMZN = AMZN.set_index('Date')

# Group data by month and calculate the mean of 'High' and 'Low'
monthly_averages = AMZN[['High', 'Low']].resample('M').mean()

# Create the line chart
plt.figure(figsize=(12, 6))
plt.plot(monthly_averages.index, monthly_averages['High'], label='Average High')
plt.plot(monthly_averages.index, monthly_averages['Low'], label='Average Low')

# Customize the chart
plt.xlabel('Month')
plt.ylabel('Price')
plt.title('Average Monthly High and Low Prices')
plt.legend()
plt.grid(True)
plt.show()
```
